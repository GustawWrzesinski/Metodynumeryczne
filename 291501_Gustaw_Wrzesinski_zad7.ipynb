```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import math as m

xw = 29
tw = 2500
dx = 1/xw
dt = 1/tw
L = 1
T = 1
D = 0.1
r = dt/pow(dx, 2)

Rys = []
WarStabil = dt/dx

if WarStabil <= 1/2:
    print("Warunek stabilnosci jest spelniony")

    # warunki poczatkowe
    u = [0 for j in range(xw+1)]
    for i in range(xw):
        x = i*dx
        u[i] = m.sin(3.14*x)
    for t in range(1,tw+1):
        uniu = [0 for l in range(xw+1)]
        for p in range(1,xw-1):
            uniu[p] = D*r*u[p-1] + (1-2*r*D)*u[p] + D*r*u[p + 1]
        u = uniu
        Rys.append(u)

    Xx = np.linspace(0, 1, 30)
    Tt = np.linspace(0, 1, 2500)
    [X, T] = np.meshgrid(Xx, Tt)

    fig = plt.figure(figsize=plt.figaspect(0.5))
    Wyk = fig.add_subplot(1, 2, 1)
    Wyk.pcolormesh(X, T, Rys, cmap='coolwarm')
    Wyk = fig.add_subplot(1, 2, 2, projection='3d')
    Wyk.plot_surface(X, T, np.array(Rys), cmap='coolwarm')
    plt.show()
else:
    print("Warunek Stabilnosci nie jest spelniony")
  
```
