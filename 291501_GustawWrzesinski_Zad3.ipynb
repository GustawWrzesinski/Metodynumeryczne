```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg 

m = np.array([[9,8,-2,2,-2],
              [7,-3,-2,7,2],
              [2,-2,2,-7,6],
              [4,8,-3,3,-1],
              [2,2,-1,1,4]])


print(m)

if all(len(row) == len(m) for row in m):
    print("Rozpatrywana macierz jest kwadratowa")
    print()

    n = len(m)
    u = ([[0]*n for x in range (n)])
    l = ([[0]*n for x in range (n)])
    
    for ii in range (n):
        u[ii][ii] = 1
        #L[ii][ii] = 1
        for kk in range (ii, n):
            suma  = sum([u[jj][kk]*l[ii][jj] for jj in range(ii)])
            u[ii][kk] = m[ii][kk] - suma
            
        for kk in range(ii, n):
            if ii == kk:
                l[ii][ii] = 1
                continue
            suma = sum([l[kk][jj]*u[jj][ii] for jj in range(ii)])
            l[kk][ii] = (m[kk][ii] -suma)/u[ii][ii]
    
    print("Macierz dolna:")
    print()
    print(np.array(l))
    print()
    print("Macierz gorna:")
    print()
    print(np.array(u))
    print()
    print("Wyniki pochodzacy z wbudowanej funkcji:")
    print()
    
    m_lu =scipy.linalg.lu(m)
    

    print(np.array(m_lu))

else:
    (print("Rozpatrywana macierz nie jest kwadratowa"))
```
